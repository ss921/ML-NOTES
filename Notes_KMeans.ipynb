## KMeans Clustering:-
* non supervised learning algorithm

```python
# we can generate data by making blobs --> is a buildin fn of sklearn
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=500, n_features=2, cluster_std=2, random_state=101, centers=4)
```

```python
# lets visualise it:
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import numpy as np
import pandas as pd

plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
plt.figure(figsize=(12,6))
plt.show()
```

```python
# applying kmeans
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4)
km.fit(data[0])
```

```python
fig, (ax1,ax2) = plt.subplots(1, 2)
ax1.set_title('Kmeans')
ax1.scatter(data[0][:,0],data[0][:,1], c=km.labels_, cmap= 'rainbow')
ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1], c=data[1], cmap= 'rainbow')
```

Anaalizing the realworld dataset --> predicting whether institution is public/private

```python
df = pd.read_csv('Python-Data-Science-and-Machine-Learning-Bootcamp-master/Machine Learning Sections/K-Means-Clustering/College_Data')
```

```python
df_feat = df.drop(labels=['Unnamed: 0','Private'], axis=1)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)

from sklearn.preprocessing import scale

kmeans.fit(scale(df_feat))
```

```python
d = {'Yes':1, 'No':0}
y = np.array(df['Private'].map(d))
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print(accuracy_score(y, kmeans.labels_))
print('\n')
print(confusion_matrix(y,kmeans.labels_))
```
